```python
# Import Dependencies
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
# from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
# Check the information of the DataFrame
df.head()
```

```python
# Convert all of the rows to float64 to get the same type for all cells
for column,cell in df.items():
    if df[column].dtype=='int64':
        df=df.astype({column:'float64'})
```

```python
# Double check the info
df.info()
```

```python
# Check the unique values in koi_disposition column
df['koi_disposition'].unique()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
X = df.drop('koi_disposition',axis=1)
y=df['koi_disposition']
print(X.shape,y.shape)
```

# Convert the koi_disposition to numerical values

```python
# Convert the string to numerical value
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
```

```python
for label, original_class in zip(encoded_y, y):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# Create the Train_Test_Split 
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.5, random_state=42)
```

```python
# Display the training set
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# We want to scale our data down to a simple numerical range. We do not want to have a big range, cause the ML will work better 
# with one range of data. In this case, I use the MinMaxScaler to scale out the X values
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Train the Model



```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=40))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=3, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

# Make Predictions

```python
encoded_predictions = model.predict_classes(X_test_scaled[:10])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
# Print out the prediction and actual labels
print(f"Predicted classes: {prediction_labels}")
print(f"Actual Labels: {list(y_test[:10])}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Pick the features for grid search models
selected_features=df[['koi_period','koi_impact','koi_depth','koi_insol','koi_kepmag']]
```

```python
# Set the X and y
data=selected_features
target=df['koi_disposition']
```

```python
# Convert the y to numerical values
label_encoder = LabelEncoder()
label_encoder.fit(target)
encoded_target = label_encoder.transform(target)
```

```python
# Create different train_split_test set
X_train, X_test, y_train, y_test = train_test_split(data, encoded_target,test_size=0.5, random_state=42)
```

```python
X_train.head()
```

```python
# Create the GridSearchCV model
# Support vector machine linear classifier
from sklearn.svm import SVC 
model = SVC(kernel='linear')
# Create the GridSearch estimator along with a parameter object containing the values to adjust
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1,5,10],
              'gamma': [0.0001, 0.001, 0.01]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
